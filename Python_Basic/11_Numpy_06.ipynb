## Numpy의 통계함수 : min, max, median, mean, std, var

#### axis 지정하지 않는다면 다차원 배열의 결과를 모두 1차원으로 모두 생각하고 연산을 진행함.

````python
import numpy as np

# 정규분포의 형태를 띄는 5개 원소를 가진 np.ndarray 타입의 배열
x = np.random.normal(3, 0.5, 5) 
print(x)
print(np.min(x))
print(np.max(x))
print(np.median(x))
print(np.mean(x))
print(np.std(x))
print(np.var(x))
````

````python
import numpy as np
x2 = np.random.normal(3, 0.5, (4,5))
print(x2)
print(np.min(x2))
print(np.max(x2))
print(np.median(x2))
print(np.mean(x2))
print(np.std(x2))
print(np.var(x2))
````

#### axis 적용시 다음과 같이 해당하는 axis 방향으로 함수 적용하여 반환.

````python
print(x2)
print(np.min(x2, axis=0))
print(np.min(x2, axis=1))
````

````python
import numpy as np
x3 = np.random.normal(3, 0.5, (4, 3, 5))
print(x3)
print(np.min(x3))  # 1차원 결과 반환.
print(np.min(x3, axis=0))
print(np.min(x3, axis=1))
print(np.min(x3, axis=2))
````

## axis 사용 후 연산 결과는 해당 axis칸의 해당하는 값 빼고 튜플 형태 반환
- ex) axis = 0 일 때 (4,3,5) 배열의 반환 shape (3,5)

````python
print(x3.shape)

# 해당 
print(np.min(x3, axis=0).shape)
print(np.min(x3, axis=1).shape)
print(np.min(x3, axis=2).shape)
````

````python
import matplotlib.pyplot as plt

from scipy import misc
face = misc.face()
plt.imshow(face)
plt.show()
````

## Numpy 에서 3차원 칼라 영상을 2차원 흑백 영상으로 합치는 방법.

````python
face_gray = np.mean(face, axis=-1)
face_gray.shape, face_gray.dtype
````

````python
plt.imshow(face_gray, cmap='gray')
plt.show()
````
