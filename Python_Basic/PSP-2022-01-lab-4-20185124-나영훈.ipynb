## 예제 3.
이 dictionary에서 각 개인의 모든 과제 점수를 문자열에서 정수형으로 바꾸고, 과제 점수의 합계와 평균을 다음과 같이 추가하라. 평균은 소숫점 셋째자리에서 반올림해야 한다.

예를 들어 첫 학생의 key:value는 다음과 같이 수정된다.
```
{'이름': '길은영', '학번': '20267377', '소속': '일본학과', '학년': '3', '과제1': 61, '과제2': 69, '과제3': 46, '과제4': 79, '과제5': 57, '과제6': 77, '과제7': 56, '과제8': 65, '합계': 510, '평균': 63.75}
```

````python
import csv
student_records = []
with open('fake_student_records.csv', mode='r', encoding='utf-8') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    for line in csv_reader:
        total = 0
        count = 0
        for key in line.keys():
            if '과제' in key:
                line[key] = eval(line[key])
                total += line[key]
                count += 1
        line["합계"] = total
        line["평균"] = round(total/count, 2)
        
        print(line)
````

## 예제 7.
'fake_student_records.csv' 파일에 의하면 모든 학생이 여덟 번의 시험을 보았다. 시험 시행 차수 별 평균과 표준 편차를 구하고, 막대 그래프로 표시하라.

$$\mu = \frac{1}{N} \sum_{n=1}^{N} x_n, \;\; \sigma = \sqrt{\frac{1}{N} \sum_{n=1}^{N} (x_n-\mu)^2} $$

아래 메쏘드를 사용할 때, yerr argument에 표준 편차를 지정하면 된다.

```
matplotlob.pyplot.bar()

```

````python
total = {f"과제{n}":0 for n in range(1,9) }

with open('fake_student_records.csv', mode='r', encoding='utf-8') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    cnt = 0
    for row in csv_reader:
        for key in row:
            cnt+=1
            if "과제" in key:
                total[key] += eval(row[key])
        
````

````python
total
````

````python
avg = { k : round(total[k]/cnt, 5) for k in total}
avg
````

````python
import math

std = {f"과제{n}":0 for n in range(1,9) }

with open('fake_student_records.csv', mode='r', encoding='utf-8') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    cnt = 0
    for row in csv_reader:
        for key in row:
            cnt+=1
            if "과제" in key:
                std[key] += (eval(row[key]) - avg[key])**2

for k in std:
    std[k] /= count
    std[k] = math.sqrt(std[k])
````
