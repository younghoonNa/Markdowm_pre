# Matplotlib

# Matplotlib.pyplot
- matplotlib가 MATLAB처럼 작동하도록 명령 스타일 함수들의 모음
- 한 문장의 명령을 추가할때 마다 figure에 변화 발생

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
x = [0,1,2,3]
y =[1,-1, 1, -1]
plt.plot(x,y)
```

```python
%matplotlib inline
x = [0,1,2,3]
y =[1,-1, 1, -1]
plt.plot(x,y)
plt.show() # 반환형함수를 호출하지 않음. 대화형 셀 에서만 필요
```

### plt.label()

```python
%matplotlib inline
x = [0,1,2,3]
y =[1,-1, 1, -1]
plt.plot(x,y)
plt.xlabel("$x$")
```

```python
%matplotlib inline
x = [0,1,2,3]
y =[1,-1, 1, -1]
plt.plot(x,y, 'go-') # g : green  , o : 점,   - : 실선
plt.xlabel("$x$")        
```

```python
%matplotlib inline
x = [0,1,2,3]
y =[1,-1, 1, -1]
plt.plot(x,y, 'r*--') # r : red  , * : 별모양 점,   -- : 점선
plt.xlabel("$x$")        
```

```python
%matplotlib inline
x = [0,1,2,3]
y =[1,-1, 1, -1]
plt.plot(x,y, 'y^--') # y : yellow  , ^ : 삼각형 모양 점,   -. : -.-. 형태의 점
plt.xlabel("$x$")        
```

```python
%matplotlib inline
x = [0,1,2,3]
y =[1,-1, 1, -1]
plt.plot(x,y, 'md:') # m : magenta  , d : 다이아몬드 모양 점,   `:` : 점이 좀 많은 선
plt.xlabel("$x$")        
```

```python
plt.plot(x,y, 'kh-', markersize= 10) #h는 6각형
plt.xlabel("$x$")
```

```python
%matplotlib inline
x = np.arange(0,1,0.01)
y = np.sin(2 * np.pi * x)
y2 = np.cos(2 * np.pi * x)

plt.plot(x,y, 'c-', x, y2, 'm-.')
plt.plot(x,y, linestyle = "-", color = "cyan", linewidth = 3)
plt.plot(x, y2, ls = '-.', color = 'm', lw = 5)
plt.xlim(0,1)
plt.ylim(0,1)
plt.xticks([0, 0.5, 1])
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.show()
```

## plt.legend()

```python
%matplotlib inline
x = np.arange(0,1,0.01)
y = np.sin(2 * np.pi * x)
y2 = np.cos(2 * np.pi * x)

plt.plot(x,y, 'c-', x, y2, 'm-.')
plt.plot(x,y, linestyle = "-", color = "cyan", linewidth = 3, 
         label = "$\sin(2\pi x)$")
plt.plot(x, y2, ls = '-.', color = 'm', lw = 5,
        label = "$\cos(2\pi x)$")
plt.xlim(0,1)
plt.ylim(0,1)
plt.xticks([0, 0.5, 1])
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.legend()
plt.show()
```

```python
%matplotlib inline
x = np.arange(0,1,0.01)
y = np.sin(2 * np.pi * x)
y2 = np.cos(2 * np.pi * x)

plt.plot(x,y, 'c-', x, y2, 'm-.')
plt.plot(x,y, linestyle = "-", color = "cyan", linewidth = 3, 
         label = "$\sin(2\pi x)$")
plt.plot(x, y2, ls = '-.', color = 'm', lw = 5,
        label = "$\cos(2\pi x)$")
plt.xlim(0,1)
plt.ylim(0,1)
plt.xticks([0, 0.5, 1])
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.legend(loc= 9)
plt.show()
```

## 그림의 종횡비 (aspect ratio) 설정

```python
t = np.linspace(np.pi/2, np.pi/2 + 4*np.pi, 6)
x = np.cos(t)
y = np.sin(t)

plt.plot(x,y)
plt.show()
```

```python
plt.gca().set_aspect(True)
t = np.linspace(np.pi/2, np.pi/2 + 4*np.pi, 6)
phi = np.pi/2
x = np.cos(t)
y = np.sin(t)

plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize = (8,3))
x = np.arange(0, 1, 0.01)
y = np.sin(2 * np.pi * x)
y2 = np.cos(2 * np.pi * x)

plt.plot(x,y, 'c-', x, y2, 'm-.')
plt.plot(x,y, linestyle = "-", color = "cyan", linewidth = 3, 
         label = "$\sin(2\pi x)$")
plt.plot(x, y2, ls = '-.', color = 'm', lw = 5,
        label = "$\cos(2\pi x)$")
plt.xlim(0,1)
plt.ylim(0,1)
plt.xticks([0, 0.5, 1])
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.legend(loc = 9)
plt.title("Plots of sine and cosine", loc = "right")


ax = plt.gca() # 이 그림에서 그래프다 그려지는 영역이 모두 표현되는 모든 axis 포함
ax.spines['right'].set_color('none') 
# spines : 바운더리 박스 조절 , spines['right'] : 바운더리 박스 중 오른쪽 박스 조절
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(("data", 0))
plt.show()
```
