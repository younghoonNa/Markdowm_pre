# 정수의 비트 단위 이동 (Bit-wise shift)

- m >> n 은 m//(2**n) 과 동일 
- m << n 은 m*(2**n) 과 동일

```python
15 >> 2, 15//2**2 , 3877 >> 5 , 3877 // 2**5
```

```python
-15 >> 2, -15//2**2, -3877 >> 5, -3877 // 2**5
```

### 항상 음수 단위의 shift 연산은 양수 나머지의 + 1 이라는걸 기억하자.

# 명령 줄과 들여쓰기

- 동일블록 내에서는 동일 블럭 사용.
- 들여쓰기 공백의 규정은 없으나 공백 4개 권장

```python
fact = False

if fact == True:
    print("Answer")
    print('True')
    
else:
        print("Answer")
        print("False")
```
